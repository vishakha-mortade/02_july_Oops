````python
1. Explain what inheritance is in object-oriented programming and why it is used.
````

````python
In object-oriented programming, inheritance is a mechanism where a class inherits the properties and methods of another class. It allows for code reuse, promotes code organization, and facilitates the creation of hierarchical relationships between classes. It is used in Python to create new classes based on existing ones, promoting code reusability and modularization.
````

````python
2. Discuss the concept of single inheritance and multiple inheritance, highlighting their
differences and advantages.
````

````python
Single inheritance is a concept where a class inherits attributes and methods from only one parent class. Multiple inheritance allows a class to inherit from multiple parent classes. The main difference is that single inheritance provides a simpler and more straightforward structure, while multiple inheritance allows for greater flexibility and code reuse. In Python, both concepts can be implemented using the "class" keyword. 
````

````python
3. Explain the terms "base class" and "derived class" in the context of inheritance.
````

````python
 In object-oriented programming, a base class is a class from which other classes inherit properties and behaviors. A derived class, also known as a subclass, is a class that derives or inherits those properties and behaviors from a base class. 
````

````python
4. What is the significance of the "protected" access modifier in inheritance? How does
it differ from "private" and "public" modifiers?
````

````python
The "protected" access modifier in Python is used to indicate that a class member should be accessible within the class and its subclasses only. It limits access to prevent direct modification or access outside of the class hierarchy while allowing subclasses to use and override the member.
Unlike "private" members, which can only be accessed within the class that defines them, "protected" members can also be accessed within subclasses.
On the other hand, "public" access modifiers have no access restrictions and can be accessed from anywhere, including outside of the class hierarchy.

````

````python
5. What is the purpose of the "super" keyword in inheritance? Provide an example.
````

````python
The "super" keyword in inheritance is used to refer to the parent class and enables us to call its methods or access its attributes. It is particularly useful when we want to override a method in the child class, but still want to utilize the parent class's implementation. Here's an example:

```python
class Parent:
def greet(self):
print("Hello from Parent")

class Child(Parent):
def greet(self):
super().greet() # calling parent class's greet method
print("Hello from Child")

child = Child()
child.greet()
```

Output:
```
Hello from Parent
Hello from Child
```

In this example, the child class `Child` overrides the `greet()` method, but by using `super().greet()`, it first calls the parent class's `greet()` method and then adds its own behavior.
````

````python
7. Create a base class called "Employee" with attributes like "name" and "salary."
Derive two classes, "Manager" and "Developer," from "Employee." Add an additional
attribute called "department" for the "Manager" class and "programming_language"
for the "Developer" class.
````

````python
``python
class Employee:
def __init__(self, name, salary):
self.name = name
self.salary = salary


class Manager(Employee):
def __init__(self, name, salary, department):
super().__init__(name, salary)
self.department = department


class Developer(Employee):
def __init__(self, name, salary, programming_language):
super().__init__(name, salary)
self.programming_language = programming_language


manager = Manager("John Doe", 5000, "Human Resources")
developer = Developer("Jane Smith", 4000, "Python")

print(f"Manager Information:")
print(f"Name: {manager.name}")
print(f"Salary: ${manager.salary}")
print(f"Department: {manager.department}")

print(f"\nDeveloper Information:")
print(f"Name: {developer.name}")
print(f"Salary: ${developer.salary}")
print(f"Programming Language: {developer.programming_language}")
```

Output:
```
Manager Information:
Name: John Doe
Salary: $5000
Department: Human Resources

Developer Information:
Name: Jane Smith
Salary: $4000
Programming Language: Python
```

In this example, the `Employee` class is the base class with attributes `name` and `salary`. The `Manager` class is derived from `Employee` and adds the `department` attribute. The `Developer` class is also derived from `Employee` and adds the `programming_language` attribute. Both derived classes use the `super()` function to call the parent class's `__init__()` method and initialize the base attributes.

````

````python
8. Design a base class called "Shape" with attributes like "colour" and "border_width."
Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
specific attributes like "length" and "width" for the "Rectangle" class and "radius" for
the "Circle" class.
````

````python
python
class Shape:
def __init__(self, colour, border_width):
self.colour = colour
self.border_width = border_width


class Rectangle(Shape):
def __init__(self, colour, border_width, length, width):
super().__init__(colour, border_width)
self.length = length
self.width = width


class Circle(Shape):
def __init__(self, colour, border_width, radius):
super().__init__(colour, border_width)
self.radius = radius


rectangle = Rectangle("Red", 2, 10, 5)
circle = Circle("Blue", 1, 7)

print(f"Rectangle Information:")
print(f"Colour: {rectangle.colour}")
print(f"Border Width: {rectangle.border_width}")
print(f"Length: {rectangle.length}")
print(f"Width: {rectangle.width}")

print(f"\nCircle Information:")
print(f"Colour: {circle.colour}")
print(f"Border Width: {circle.border_width}")
print(f"Radius: {circle.radius}")
```

Output:
```
Rectangle Information:
Colour: Red
Border Width: 2
Length: 10
Width: 5

Circle Information:
Colour: Blue
Border Width: 1
Radius: 7
```

In this example, the `Shape` class is the base class with attributes `colour` and `border_width`. The `Rectangle` class is derived from `Shape` and adds the `length` and `width` attributes. The `Circle` class is also derived from `Shape` and adds the `radius` attribute. Both derived classes use the `super()` function to call the parent class's `__init__()` method and initialize the base attributes.

````

````python
10. Create a base class called "BankAccount" with attributes like "account_number" and
"balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
"BankAccount." Add specific methods like "calculate_interest" for the
"SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.
````

````python
class BankAccount:
def __init__(self, account_number, balance):
self.account_number = account_number
self.balance = balance


class SavingsAccount(BankAccount):
def __init__(self, account_number, balance):
super().__init__(account_number, balance)

def calculate_interest(self, rate):
interest = self.balance * rate
self.balance += interest


class CheckingAccount(BankAccount):
def __init__(self, account_number, balance):
super().__init__(account_number, balance)

def deduct_fees(self, fee):
self.balance -= fee


savings = SavingsAccount("123456789", 5000)
checking = CheckingAccount("987654321", 3000)

savings.calculate_interest(0.05)
checking.deduct_fees(50)

print(f"Savings Account Information:")
print(f"Account Number: {savings.account_number}")
print(f"Balance: ${savings.balance}")

print(f"\nChecking Account Information:")
print(f"Account Number: {checking.account_number}")
print(f"Balance: ${checking.balance}")
```

Output:
```
Savings Account Information:
Account Number: 123456789
Balance: $5250.0

Checking Account Information:
Account Number: 987654321
Balance: $2950
```

In this example, the `BankAccount` class is the base class with attributes `account_number` and `balance`. The `SavingsAccount` class is derived from `BankAccount` and adds the `calculate_interest` method to calculate and add interest to the account balance. The `CheckingAccount` class is also derived from `BankAccount` and adds the `deduct_fees` method to deduct fees from the account balance. Both derived classes use the `super()` function to call the parent class's `__init__()` method and initialize the base attributes.

````
